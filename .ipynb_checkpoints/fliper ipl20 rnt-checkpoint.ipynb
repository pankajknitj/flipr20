### Pipeline

```python
selected_col
```

```python
class model:
    def __init__(self,selected_col):
        self.feature=np.nan
        self.feature_scaled=np.nan
        self.selected_feature=selected_col
    def mining(self):
        self.feature.drop(['PLAYER','Mat'],axis=1,inplace=True)
        self.feature.HS=self.feature['HS'].map(str).str.replace('*','').astype(int)

    def feature_gen(self):
        self.feature['Runs_by_boundry']=self.feature['4s']*4+6*self.feature['6s']
        self.feature['%Runs by boundry']=np.round(self.feature.Runs_by_boundry/self.feature['prev_Runs']*100,2)
        self.feature['Avg_b']=np.round(self.feature.Runs_by_boundry/self.feature.Inns,2)
        self.feature['bat_avg'] = self.feature.apply(lambda x: (x['prev_Runs']/x['Inns']) if x.Inns==x.NO else (x['prev_Runs']/(x['Inns']-x['NO'])), 
                    axis = 1)
        self.feature['bat_avg'] = self.feature['bat_avg'].apply(lambda x: np.round(x,2))
    def predict(self,f):
        self.feature=f
        self.mining()
        self.feature_gen()
        self.feature=self.feature[self.selected_feature]
        self.feature_scaled=ss.transform(self.feature)
        #print(self.feature.columns)
        return np.floor(lgb_cl.predict(self.feature)*.5+lr.predict(feature_scaled)*.5)
```
